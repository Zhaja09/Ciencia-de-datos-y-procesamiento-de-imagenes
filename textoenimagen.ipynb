```python
import cv2
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow


img = cv2.imread("/content/img3.png", cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
hist = cv2.calcHist([img], [0], None, [256], [0,256])
plt.plot(hist, color="gray")
plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()
```

```python
import cv2
from skimage import io

def texto():

    fonts = [cv2.FONT_ITALIC, cv2.FONT_HERSHEY_DUPLEX, cv2.FONT_HERSHEY_SCRIPT_COMPLEX, cv2.FONT_ITALIC]
    color = [(0,255,0),(255,0,0),(0,0,255),(0,0,0)]
    print ("seleccione el tipo de letra")
    for i, font in enumerate(fonts, start=1):
      print (f"{i}, {font}")
    font_E = int(input("seleccione un numero del 1 al 4 para seleccionar la fuente"))
    if font_E < 1 or font_E > len(fonts):
        print("Eleccin no vlida. Por favor, intente de nuevo.")
        return

    print("Seleccione un color:")
    for i, color in enumerate(color, start=1):
        print(f"{i}. {color}")

    color_E = int(input("Ingrese el nmero correspondiente a su eleccin: "))

    if  color_E < 1 or color_E > len (color):
      print ("eleccion no valida")
      return

    cordenadas = input("ingrese las coordenadoas")
    try:
      x,y = map (int , cordenadas.split(","))
    except ValueError:
      print("coordenadas no validas")
      return

    print(f"Tipo de letra seleccionada:  {fonts[font_E - 1]}")
    print(f"color de letra seleccionado:  {color[color_E - 1]}")
    print(f"coordenadas seleccionadas:  ({x},{y})")

    imagen = cv2.imread("/content/FKRiBHOWYAEI8i-.jpg")

    cv2.putText(imagen, "prueba texto",(x,y), fonts[font_E - 1], 1.9, color[color_E - 1], 4, cv2.LINE_AA)
    io.imshow(imagen)

texto()
```
