```python
def estan_balanceados(cadena):
    pila = []
    mapeo = {")": "(", "]": "[", "}": "{"} # Mapeo de símbolos de cierre a apertura
    simbolos_apertura = set("([{")
    simbolos_cierre = set(")]}")

    for char in cadena:
        if char in simbolos_apertura:
            pila.append(char)
        elif char in simbolos_cierre:
            if not pila: # Si la pila está vacía, hay un símbolo de cierre sin apertura
                return False
            top_elemento = pila.pop()
            if mapeo[char] != top_elemento: # Si el símbolo de cierre no coincide con el último de apertura
                return False

    return not pila # Si la pila está vacía al final, todo está balanceado

# Ejemplos de uso:
print(f"'({[]})' está balanceado: {estan_balanceados('({[]})')}")
print(f"'([)]' está balanceado: {estan_balanceados('([)]')}")
print(f"'{{[()]}}' está balanceado: {estan_balanceados('{[()]}}')}")
print(f"'{{{{[]}}}}' está balanceado: {estan_balanceados('{{{{[]}}}}')}")
print(f"'((()))' está balanceado: {estan_balanceados('((()))')}")
print(f"'())' está balanceado: {estan_balanceados('())')}")
print(f"'' está balanceado: {estan_balanceados('')}")
```

```python
def evaluar_rpn(expresion):
    pila = []
    tokens = expresion.split()

    for token in tokens:
        if token.isdigit() or (token[0] == '-' and token[1:].isdigit()): # Check if it's a number (positive or negative)
            pila.append(float(token)) # Convert to float to handle division results
        else:
            # It's an operator, so pop two operands
            if len(pila) < 2:
                raise ValueError("Expresión RPN inválida: faltan operandos.")

            operando2 = pila.pop()
            operando1 = pila.pop()

            if token == '+':
                pila.append(operando1 + operando2)
            elif token == '-':
                pila.append(operando1 - operando2)
            elif token == '*':
                pila.append(operando1 * operando2)
            elif token == '/':
                if operando2 == 0:
                    raise ZeroDivisionError("Error: División por cero.")
                pila.append(operando1 / operando2)
            else:
                raise ValueError(f"Operador desconocido: {token}")

    if len(pila) != 1:
        raise ValueError("Expresión RPN inválida: demasiados operandos.")

    return pila[0]

# Ejemplos de uso:
print(f"'3 4 + 2 *' = {evaluar_rpn('3 4 + 2 *')}") # Salida esperada: 14.0
print(f"'10 5 / 3 +' = {evaluar_rpn('10 5 / 3 +')}") # Salida esperada: 5.0
print(f"'1 2 + 3 4 + *' = {evaluar_rpn('1 2 + 3 4 + *')}") # Salida esperada: 21.0
print(f"'5' = {evaluar_rpn('5')}") # Salida esperada: 5.0
print(f"'7 2 -' = {evaluar_rpn('7 2 -')}") # Salida esperada: 5.0
print(f"'8 2 3 * +' = {evaluar_rpn('8 2 3 * +')}") # Salida esperada: 14.0 (8 + (2*3))
```
