```python
from skimage import io
from skimage import color
import pylab as pp
imagen = io.imread("/content/ejemplo1.jpg")
def conversionYuv(imagen):
    imagen_Yuv = color.rgb2yuv(imagen)
    imagen_rgb = color.lab2rgb(imagen_Yuv)
    pp.figure(4)
    pp.title("imagen_Yuv")
    io.imshow(imagen_Yuv)
    pp.figure(5)
    io.imshow(imagen_rgb)

conversionYuv(imagen)
```
