```python
from google.colab.patches import cv2_imshow
import cv2
from pylab import *
import matplotlib.pyplot as pp

imagen_Original = cv2.imread("/content/luz.jpg")

#Parameters : src : Imagen
#kernel size: Tupla tamao borroso
#dst It is the output image of the same size and type as src.
#return value: It returns an image.

image_media = cv2.blur (imagen_Original,(4,6))
image_medianBlur=cv2.medianBlur(imagen_Original, 9)
image_GaussianBlur=cv2.GaussianBlur(imagen_Original,(9,9),10)
image_BilateralBlur=cv2.bilateralFilter(imagen_Original,7,152,45)
#show images
figure(0)
cv2_imshow(imagen_Original)

figure(1)
cv2_imshow(image_media)

figure(2)
cv2_imshow(image_medianBlur)

figure(3)
cv2_imshow(image_GaussianBlur)

figure(4)
cv2_imshow(image_BilateralBlur)
```
