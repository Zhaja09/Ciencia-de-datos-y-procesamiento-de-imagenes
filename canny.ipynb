```python

import numpy as np
from google.colab.patches import cv2
import cv2
from google.colab.patches import cv2_imshow
from skimage import io

def UsoCanny():
  imagen = io.imread("/content/objetos-cotidianos.jpg")
  gris  = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
  imagenSuavizada = cv2.GaussianBlur(gris, (3, 3), 0)
  canny = cv2.Canny(imagenSuavizada, 200, 100)
  cv2_imshow(canny)
  (contornos, _) = cv2.findContours(canny, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
  print("posibles objetos encontrados {} objetos".format(len(contornos)))
  cv2.drawContours(imagen, contornos, -1, (0, 255, 0), 2)
  cv2_imshow(imagen)
UsoCanny()
```
