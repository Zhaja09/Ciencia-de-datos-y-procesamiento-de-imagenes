```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear un grafo dirigido
G = nx.DiGraph()

# Aadir nodos para las entradas y las puertas
G.add_node("A", pos=(0, 2))
G.add_node("B", pos=(0, 1))
G.add_node("C", pos=(0, 0))
G.add_node("NOT A", pos=(2, 2))
G.add_node("NOT B", pos=(2, 1))
G.add_node("NOT C", pos=(2, 0))
G.add_node("AND1", pos=(4, 1.5))
G.add_node("AND2", pos=(4, 0.5))
G.add_node("AND3", pos=(4, 0))
G.add_node("OR", pos=(6, 1))

# Aadir aristas para las conexiones entre nodos
G.add_edges_from([("A", "NOT A"), ("B", "AND1"), ("C", "AND1"),
                  ("A", "AND2"), ("NOT B", "AND2"), ("C", "AND2"),
                  ("A", "AND3"), ("B", "AND3"), ("NOT C", "AND3"),
                  ("B", "NOT B"), ("C", "NOT C")])

G.add_edges_from([("NOT A", "AND1"), ("NOT B", "AND2"), ("NOT C", "AND3"),
                  ("AND1", "OR"), ("AND2", "OR"), ("AND3", "OR")])

# Obtener la posicin de los nodos
pos = nx.get_node_attributes(G, 'pos')

# Dibujar el grafo
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color='lightblue', font_size=10, font_weight='bold', arrows=True)
plt.title("Diagrama del Circuito Digital")
plt.show()
```
